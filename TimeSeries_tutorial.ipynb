### Forecasting temperatures

```python
import tensorflow as tf
import os
import pandas as pd
import numpy as np
```

```python
datasetURL = 'https://raw.githubusercontent.com/gilbutITbook/006975/master/datasets/jena_climate/jena_climate_2009_2016.csv' #link from github
df = pd.read_csv(datasetURL)
```

```python
df #every 10 minutes a measure.
```

```python
df = df[5::6] #every hour
df
```

```python
df.index = pd.to_datetime(df['Date Time'], format = '%d.%m.%Y %H:%M:%S')
df[:26]
```

```python
temp = df['T (degC)'] #temperature
temp.plot()
```

```python
#formatting data for the model, create sequence and output. supervised learning
#something like 
#[[[1],[2],[3],[4],[5]]] [6]
#[[[2],[3],[4],[5],[6]]] [7]
#every variable is a list so it may be something like: [temp, pressure, etc...], this is multivariable

def df_to_X_y(df, window_size=5):
  df_as_np = df.to_numpy() #to iterate over it
  X = []
  y = []
  for i in range(len(df_as_np)-window_size):
    row = [[a] for a in df_as_np[i:i+5]] #[a] to do the trick for lists
    X.append(row)
    label = df_as_np[i+5]
    y.append(label)
  return np.array(X), np.array(y)
```

```python
WINDOW_SIZE = 5
X, y = df_to_X_y(temp, WINDOW_SIZE)
X.shape, y.shape
```

```python
import sklearn
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(
    X, y, test_size=0.1, random_state=0)

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam


model = Sequential()
model.add(InputLayer((5,1)))
model.add(LSTM(64))
model.add(Dense(8, 'relu'))
model.add(Dense(1, 'linear')) #final temp

model.summary()
```

```python
cp = ModelCheckpoint('model/', save_best_only=True) #only save the best model
model.compile(loss = MeanSquaredError(), optimizer = Adam(learning_rate=0.0001), metrics = [RootMeanSquaredError()])
```

```python
model.fit(X_train, y_train, epochs = 10, callbacks=[cp])
```

```python
train_predictions = model.predict(X_train).flatten()
train_results = pd.DataFrame(data={'Train Predictions': train_predictions, 'Actuals': y_train})
train_results
```

```python
test_predictions = model.predict(X_test).flatten()
test_results = pd.DataFrame(data={'Train Predictions': test_predictions, 'Actuals': y_test})
test_results
```

```python
import matplotlib.pyplot as plt
plt.plot(test_results['Train Predictions'][:50])
plt.plot(test_results['Actuals'][:50])
```
